```python
import numpy as np


def mc_pitts_andnot(x, y):
    
    w1 = 1   
    w2 = -1  
    threshold = 1
    
    
    net_input = w1 * x + w2 * y
    
    
    return 1 if net_input >= threshold else 0


inputs = [(0,0), (0,1), (1,0), (1,1)]

print("X Y | ANDNOT")
print("------------")
for x, y in inputs:
    print(f"{x} {y} |   {mc_pitts_andnot(x, y)}")
```
