```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])

w=[0,0]
b=0
lr=0.1

def activation(x):
    return 1 if x>=0 else 0
for epoch in range(20):
    for i in range(len(X)):
        x1=X[i][0]
        x2=X[i][1]
        output=x1*w[0]+x2*w[1]+b
        y_pred=activation(output)
        error=y[i]-y_pred
        w[0]=w[0]+lr*error*x1
        w[1]=w[1]+lr*error*x2
        b=b+lr*error
print("weight : ",w)
print("bias : ",b)
x=np.linspace(-1,2,100)
y_line=-(w[0]*x+b)/w[1]
for i in range(len(X)):
    if y[i]==0:
        plt.scatter(X[i][0],X[i][1])
    else:
        plt.scatter(X[i][0],X[i][1])


plt.plot(x,y_line)
plt.title("perceptron distribution region")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
plt.show()
```
