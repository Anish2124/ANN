```python
import numpy as np

X1=np.array([1,1,-1])
Y1=np.array([1,-1])

X2=np.array([-1,1,1])
Y2=np.array([-1,1])

W=np.outer(X1,Y1)+np.outer(X2,Y2)

print("weighted matrix : ")
print(W)

x=np.array([1,1,-1])
y=np.dot(x,W)

for i in range(len(y)):
    if y[i]>=0:
        y[i]=1
    else:
        y[i]=-1
print("vector")
print(y)

x_new=np.dot(W,y)

for i in range(len(x_new)):
    if x_new[i]>=0:
        x_new[i]=1
    else:
        x_new[i]=-1
print("recalled vector")
print(x_new)
```
